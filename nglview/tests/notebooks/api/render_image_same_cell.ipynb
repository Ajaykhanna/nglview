```python
import nglview as nv
import threading
import time
from ipywidgets import Widget
import pytraj as pt
from nglview.contrib.movie import MovieMaker

v = nv.demo()
v
```

```python
data = []

def render():
    v.clear()
    v.add_rope()
    s0 = v.render_image()
    v.clear()
    v.add_cartoon()
    s1 = v.render_image()
    data.extend([s0, s1])
    
render()
```

```python
assert data[0].value != data[1].value
```
